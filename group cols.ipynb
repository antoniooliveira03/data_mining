```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('data/data_copy_no_modification.csv')
```

```python
data
```

**Customer Information**

```python
customer_info = [
    'customer_id', 'customer_region', 
    'customer_age', 'is_repeat_customer'
]
```

**Temporal Data:**

```python
temporal_data = [
    'first_order', 'last_order', 'last_promo', 'days_between',
    *['HR_' + str(i) for i in range(24)], 
    *['DOW_' + str(i) for i in range(7)], 
    '0_7h', '8_14h', '15_19h', '20_23h'
]
```

**Product and Vendor Interaction**

```python
product_vendor = [
    'vendor_count', 'product_count', 'is_chain'
]
```

**Spending and Orders**

```python
spending_orders = [
    'total_orders', 'total_spend', 'avg_spend_prod',
    'promo_DELIVERY', 'promo_DISCOUNT', 'promo_FREEBIE', 'promo_NO DISCOUNT',
    'payment_method', 'pay_CARD', 'pay_CASH', 'pay_DIGI', 
    'payment_method_enc', 'last_promo_enc'
]
```

**Cuisine Preferences**

```python
cuisine_preferences = [
    col for col in data.columns if col.startswith('CUI_')
]
```

Save as a dict

```python
column_groups = {
    'Customer Information': customer_info,
    'Temporal Data': temporal_data,
    'Product and Vendor Interaction': product_vendor,
    'Spending and Orders': spending_orders,
    'Cuisine Preferences': cuisine_preferences
}
```

```python
data[customer_info].to_csv('data/customer_info.csv')
data[temporal_data].to_csv('data/temporal_data.csv')
data[product_vendor].to_csv('data/product_vendor.csv')
data[spending_orders].to_csv('data/spending_orders.csv')
data[cuisine_preferences].to_csv('data/cuisine_preferences.csv')
```
